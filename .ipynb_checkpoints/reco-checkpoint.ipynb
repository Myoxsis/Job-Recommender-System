```python
import pandas as pd
import re
```

```python
df = pd.read_csv('offers_test.csv')
```

```python
df.head()
```

```python
# Remove Chinese offers
for x in df['name']:
    if len(re.findall(r'[\u4e00-\u9fff]+', x)) > 0:
        print(x)
```

```python
i = 0
while i < len(df)-1:
    x = df.iloc[i, :]
    print(x)
    i = i + 1
```

```python
df.iloc[1, :]
```

```python
df['raw_text'] = df['Position'] + " " + df['Company'] + " "  + df['Location'] + " "  + df['desc']
```

```python
marque_1 = []

x = 0
while x <= len(df['Position'])-1:
    ft3 = ''
    ft3 = df['raw_text'][x]
    c = re.findall('logistique|supply', ft3.lower())#[0].title())
    if not c:
        marque_1.append(c)
    else:
        marque_1.append(c[0].title())
        
    x = x +1
    
df['LSCM'] = marque_1
```
